```python
import sympy

# функция для расчета последовательности
def func(n):
    return n / (sympy.factorial(n))**(1/n)
```

```python
%%time

# Поиск предела
e = 1.0e-7
i = 1
lim = func(i)

while True:
    print(f'f({i}) = {lim}')
    lim_old = lim
    i += 1
    lim = func(i)
    if abs(lim - lim_old) < e:
        break

print(f"Предел с точностью e = {e} равен {lim}")

    
```

# вариант расчета через логарифмы, чтобы без sympy

```python
import numpy

# вариант функции с расчетом через логарифмы, чтобы без sympy
def func2(n):
    # расчет факториала
    res = 0
    for i in range(1, n + 1):
        res = res + numpy.log(i)
    # извлечение корня n-степени
    res = res / n
    return n / numpy.exp(res)
```

```python
%%time

e = 1.0e-7
i = 1
lim = func2(i)

while True:
    print(f'f({i}) = {lim}')
    lim_old = lim
    i += 1
    lim = func2(i)
    if abs(lim - lim_old) < e:
        break

print(f"Предел с точностью e = {e} равен {lim}")
```
