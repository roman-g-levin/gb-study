```python
from math import sqrt

def a(n):
    return 2**n - n

def b(n):
    return 1/(1 - n)

def c(n):
    return (-1)**n + sqrt(2 * n)

def d(n):
    return (-1)**(2 * n) + 1/(n**2)
```

```python
# Исследование последовательностей
print('a(n): ', end='')
for n in range(1,20):
    print(a(n), end=' ')
print('\nМонотонна, ограничена снизу\n')

print('b(n): ', end='')
for n in range(2,20):
    print(b(n), end=' ')
print('\nМонотонна, ограничена сверху\n')

print('с(n): ', end='')
for n in range(1,20):
    print(c(n), end=' ')
print('\nНе монотонна, ограничена снизу\n')

print('d(n): ', end='')
for n in range(1,20):
    print(d(n), end=' ')
print('\nМонотонна, ограничена сверху\n\n')
```

```python
# расчет 5-тых элементов последовательностей

print(f'a(5) = {a(5)}')
print(f'b(6) = {b(6)}')
print(f'c(5) = {c(5)}')
print(f'd(5) = {d(5)}')
```
