```python
from math import factorial, pow

# функция для расчета последовательности
def func(n):
    return n / (factorial(n))**(1/n)
```

```python
# Поиск предела
e = 10e-7
i = 1
lim = func(i)

while True:
    print(f'f({i}) = {lim}')
    lim_old = lim
    i += 1
    lim = func(i)
    if abs(lim - lim_old) < e:
        break

print(f"Предел с точностью e = {e} равен {lim}")

    
```
