```python
import json
from pymongo import MongoClient
```

```python
# Подключение к серверу MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Выбор базы данных и коллекции
db = client['Books']
collection = db['inStock']
```

```python
# Вывести книги с количеством менее 3
query = {"Availability": {"$lt": 3}}
projection = {"Name": 1, "Availability": 1, "_id": 0}
selected_books = collection.find(query, projection)
for item in selected_books:
    print(item['Name'], f', Кол-во: {item['Availability']}')
```
